```python
high = [] #    
low = [] #  
for i in ls:
    if i[-1] != '' and i[-2] != '': #  csv   -1      
        if 1983 <= int(i[0].split('-')[0]): #  csv 0    0 1983     split()   
            if (i[0].split('-')[1] == '02') and i[0].split('-')[2] == '14':# csv 0    1   02   2 14 
                high.append(float(i[-1])) #  -1  high 
                low.append(float(i[-2])) #  -2  low 
```

```python
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('my birthday')
plt.plot(high, 'hotpink', label='high')
plt.plot(low, 'skyblue', label='low')
plt.legend()
```
